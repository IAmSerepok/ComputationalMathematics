```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('otvet.csv')
```

```python
plt.figure(figsize=(6, 8))

plt.subplot(2, 1, 1)
plt.plot(df['n'], df['delta'], color='lightgreen')
plt.title('     ')
plt.gca().set_yticklabels([str(i) + 'e10' for i in range(-1, 7)])


plt.subplot(2, 1, 2)
plt.plot(df['n'], df['r'], color='skyblue')
plt.title('     ')

plt.tight_layout()

plt.savefig('.jpeg', dpi=300)

plt.show()
```
