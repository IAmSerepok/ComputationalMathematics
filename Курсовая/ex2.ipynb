```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def norm_1(x):
    return np.sum(np.abs(x))
```

```python
def norm_2(x):
    return np.sqrt(np.sum(x ** 2))
```

```python
def norm_inf(x):
    return np.max(np.abs(x))
```

```python
norms = [[], [], []]
N = []
index = []
```

```python
for n in range(3, 14):
    for j in range(10):
        vec = 10 * np.random.rand(n) - 5

        N.append(n)
        index.append(10 * (n - 3) + j + 1)

        norms[0].append(norm_1(vec))
        norms[1].append(norm_2(vec))
        norms[2].append(norm_inf(vec))
```

```python
plt.figure(figsize=(14, 8))

plt.subplot(2, 2, 1)

plt.plot(index, norms[0], label='$\\|x\\|_1$')
plt.plot(index, norms[2], label='$\\|x\\|_\\infty$')
plt.plot(index, np.array(N) * np.array(norms[2]), label='$n\cdot\\|x\\|_\\infty$')

plt.legend(fontsize=16)

plt.subplot(2, 2, 2)

plt.plot(index, norms[1], label='$\\|x\\|_2$')
plt.plot(index, norms[2], label='$\\|x\\|_\\infty$')
plt.plot(index, np.array(np.sqrt(N)) * np.array(norms[2]), label='$\\sqrt{n}\\cdot\\|x\\|_\\infty$')

plt.legend(fontsize=16)

plt.subplot(2, 2, 3)

plt.plot(index, norms[0], label='$\\|x\\|_1$')
plt.plot(index, norms[1], label='$\\|x\\|_2$')
plt.plot(index, np.array(np.sqrt(N)) * np.array(norms[1]), label='$\\sqrt{n}\\cdot\\|x\\|_2$')

plt.legend(fontsize=16)

plt.subplot(2, 2, 4)

plt.plot(index, norms[0], label='$\\|x\\|_1$')
plt.plot(index, norms[1], label='$\\|x\\|_2$')
plt.plot(index, norms[2], label='$\\|x\\|_\\infty$')

plt.legend(fontsize=16)

plt.savefig('ex2.png', dpi=300, bbox_inches='tight')

plt.show()
```
