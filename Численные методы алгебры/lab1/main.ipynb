```python
import numpy as np
import pandas as pd

from random import random
```

```python
def generate_matrix(width, height):
    matrix = np.zeros((width, height), float)
    for x in range(width):
        for y in range(height):
            matrix[x, y] = random()
    return matrix
```

```python
matrix_list = [generate_matrix(25, 25) for _ in range(5)]

for _ in range(5):
    print('-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=')
    print(f'-=-=-=-=-=-=-=-=-=-= Матрица номер {_ + 1} -=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-')
    print('-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=')
    print(pd.DataFrame(matrix_list[_]))
```

```python
def matrix_norm(matrix):
    width, height = matrix.shape

    if width != height:
        raise ValueError()
    
    result = 0
    for _1 in range(width):
        result += matrix[_1][_1] ** 2
    result = np.sqrt(result) / width

    return result
```

```python
norm_list, inv_norm_list, cond_list = [], [], []

for matrix in matrix_list:
    norm = matrix_norm(matrix)
    norm_list.append(norm)

    inv_norm = matrix_norm(np.linalg.inv(matrix))
    inv_norm_list.append(inv_norm)

    cond = norm * inv_norm
    print()
    cond_list.append(cond)

df = pd.DataFrame({"Норма матрицы" : norm_list, 
                   "Норма обратной матрицы": inv_norm_list, 
                   "Показатель обусловленности": cond_list}, 
                   index=[_ for _ in range(len(matrix_list))])

df
```

```python
n = 25
a, b = -10, 10

X = np.linspace(a, b, n)

A = np.zeros((n, n), float)

for i in range(n):
    for j in range(n):
        A[i, j] = X[j] ** i

b = A @ np.array([1 for _ in range(n)])

a = np.linalg.solve(A, b)
a
```

```python
print(f'Число обусловленности матрицы Вандермонда = {matrix_norm(A) * matrix_norm(np.linalg.inv(A))}')
```

```python
def vector_norm_1(vector):
    size = vector.shape[0]

    result = 0

    for x in range(size):
        result += np.abs(x)

    return result
```

```python
print(f'Векторная норма между точным и численным решениями = {vector_norm_1(a - np.array([1 for _ in range(n)]))}')
```
